# 「ガウス過程と機械学習」
# P.68 ガウス過程からのサンプル

```julia
using Distributions
using Plots
```

### 初期値を設定

```julia
x1 = 1.0:1.0:4.0
x2 = 1.0:0.5:4.0
x3 = 1.0:0.2:4.0

θ₁ = 1.0
θ₂ = 1.0;
```

### カーネル行列を計算

```julia
# ガウスカーネル
function kernel(x, θ₁, θ₂)
    θ₁ .* exp.(-(x .- x').^2 ./ θ₂)
end
```

```julia
k1 = kernel(x1, θ₁, θ₂)
k2 = kernel(x2, θ₁, θ₂)
k3 = kernel(x3, θ₁, θ₂);
```

### ガウス過程からサンプルを生成

```julia
# それぞれ次元の数だけμを設定
μ1 = zeros(4)
μ2 = zeros(7)
μ3 = zeros(16);
```

```julia
gp1 = MvNormal(μ1, k1)
gp2 = MvNormal(μ2, k2)
gp3 = MvNormal(μ3, k3)

sample1 = rand(gp1, 1)
sample2 = rand(gp2, 1)
sample3 = rand(gp3, 1);
```

```julia
plot(
    scatter(x1, sample1, xlab="x", ylab="y", label="x -> 1.0:1.0:4.0"), 
    scatter(x2, sample2, xlab="x", ylab="y", label="x -> 1.0:0.5:4.0"), 
    scatter(x3, sample3, xlab="x", ylab="y", label="x -> 1.0:0.2:4.0")
)
```
