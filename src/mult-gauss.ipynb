```julia
using Distributions
using LinearAlgebra
using Plots
```

## 式から計算した場合

```julia
function 𝒩(x...; μ, Σ)
    Λ = inv(Σ)
    D = length(μ)
    return √(det(Λ)) * exp(-0.5 * (x .- μ)' * Λ * (x .- μ)) / √(2π)^D
end
```

```julia
μ = [
    0.0
    0.0
]

Σ = [
    1.0 0.0
    0.0 1.0
]

x1 = x2 =  -4.0:0.1:4.0
z = 𝒩.(x1', x2; μ, Σ)
contour(x1, x2, z, xlab="x1", ylab="x2")
```

## Distributions.jiを使う場合

```julia
contour(-4:0.1:4, -4:0.1:4, (x, y) -> pdf(MvNormal(μ, Σ), [x, y]), xlabel="x1", ylabel="x2")
```
