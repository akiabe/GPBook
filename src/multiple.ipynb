# P.19 重回帰

```julia
using LinearAlgebra
using Plots
```

```julia
X = [
    1 1 2
    1 -1 1
    1 3 0
    1 -2 -2
]

y = [
    4
    2
    1
    -1
]

scatter(X[:, 2], X[:, 3], y)
```

### 重回帰モデルの解

```julia
w = inv(X' * X) * X' * y
```

```julia
f(x, y) = w[1] + w[2] * x + w[3] * y
xs = -4:0.5:4
ys = -4:0.5:4
zs = f.(xs, ys)
plot(xs, ys, zs, label="Multiple Regression")
scatter!(X[:, 2], X[:, 3], y, label="Observed data")
```
