```julia
using Distributions
using Plots
```

## 式から計算した場合

```julia
function gauss(x, μ, σ)
    cons = 1.0 / √(2 * π * σ)
    cons * exp.(-(xs .- μ).^2 ./ (2 * σ^2))
end
```

```julia
xs = range(-10, 10, length=1000)
μ = 0.0
σ = 1.0

prob = gauss(xs, μ, σ)
plot(xs, prob, label="μ=0.0,σ=1.0")
```
